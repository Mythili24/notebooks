```python
import pandas as pd
df = pd.read_csv(r'Desktop\all-weeks-global.csv')
df
```

```python
df['Month'] = pd.to_datetime(df['week']).dt.month
df.head(10)
```

```python
import matplotlib.pyplot as plt
category = df.groupby(['Month', 'category']).sum()['weekly_hours_viewed']
category.unstack(level=0).plot(kind='bar')
plt.xlabel('category monthwise')
plt.ylabel('weekly_hours_viewed in crores')
plt.grid()
plt.show()
```

```python
df_films = df[df['category'].str.startswith('Films')]
film_title = df_films[['show_title', 'weekly_hours_viewed']]
film_title
```

```python
films_grp = film_title.groupby('show_title').sum()['weekly_hours_viewed'].reset_index()
films_grp
```

```python
import matplotlib.pyplot as plt
top10 = films_grp.sort_values(by=['weekly_hours_viewed'], ascending=False).head(10)
plt.bar(top10)
plt.show()
```

```python
title_list = df.groupby('show_title')
title = [x for x,y in title_list]
title
```

```python
import matplotlib.pyplot as plt
ax = plt.gca()
top10.plot(kind='line', x = 'show_title', y = 'weekly_hours_viewed', color='Red', ax = ax)
plt.show()
```
