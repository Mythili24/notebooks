```python
# Pie chart to show dismissal kind of any batsman
df_ponting = ball[ball['batsman']=='RT Ponting']
df_ponting.head()
```

```python
df_ponting['dismissal_kind'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True, rotatelabels=True)
plt.title("Dismissal Kind",fontweight="bold",fontsize=12)
plt.show()
```

```python
# To count the number of sixes and fours scored by any batsman
```

```python
def boundary(df_ponting,runs):
    return len(df_ponting[df_ponting['batsman_runs'] == runs])
```

```python
print('Number of Fours scored:', boundary(df_ponting,4))
print('Number of Sixes scored:', boundary(df_ponting,6))
```
